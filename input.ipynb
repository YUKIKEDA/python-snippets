input() 関数は、Pythonの組み込み関数です。構文は次のようになります。
```
input（[プロンプト] )
```
引数の[プロンプト]に、文字列を指定すると、入力フィールドの前にその文字列が表示されます。

<br>
【動作】

prompt 引数がある場合は，末尾の改行を除いて標準出力に書き込まれます。この関数は，入力から**1行**を読み込み，それを**文字列**に変換して（末尾の改行を取り除いて）返します。EOFを読み込んだ場合、EOFErrorが発生します。

(Ex. 1)

````python
input_line = input()
````

(Ex. 2)

````python
input_line = input("文字列を入力してください --> ")
````
